# Task 3
### Tabulate numbers. Write a program that tabulates the squares and third power of the number form 0 to 10. Try to format the tabulate nicely

```python
print("-----------------------------------------")
print(f"{"x":>5}    {"x^2":>10} {"x^3":>15}")
for i in range(0, 11):
    ii = i * i
    iii = i * i * i

    print(f"{i:>5} {ii:>10} {iii:>15}")

print("-----------------------------------------")
```
